```python
#w4a6
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
#w4a9
n = int(input("Nhap n: "))
ket_qua = []
i = 1
while i * i <= n:
    x = i * i
    if len(str(x)) == len(set(str(x))):
        ket_qua.append(x)
    i += 1

if len(ket_qua) == 0:
    print("(no number)")
else:
    print(*ket_qua)

```

```python
#w4a11
n = int(input("Nhap n: "))
if n % 2 != 0:
    print(0)
else:
    m = n // 2
    dem = 0
    i = 1
    while i * i <= m:
        if m % i == 0:
            dem += 1
            if i * i != m:
                dem += 1
        i += 1
    print(dem)
```

```python
#w4a13
def tong_uoc(n):
    s = 0
    i = 1
    while i * i <= n:
        if n % i == 0:
            if i != n:
                s += i
            if i != 1 and i * i != n:
                if n // i != n:
                    s += n // i
        i += 1
    return s

a = int(input("Nhap so thu nhat: "))
b = int(input("Nhap so thu hai: "))

if tong_uoc(a) == b and tong_uoc(b) == a:
    print("true")
else:
    print("false")
```
