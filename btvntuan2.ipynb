```python
#bài 3
a = float(input("Nhap canh a: "))
b = float(input("Nhap cạnh b: "))
c = float(input("Nhap canh c: "))
if a <= 0 or b <= 0 or c <= 0 or (a + b <= c) or (a + c <= b) or (b + c <= a):
    print("Khong la ba canh của tam giac")
else:
    print("La ba canh cua tam giac")
    if a == b == c:
        print("Tam giac deu")
    elif a == b or a == c or b == c:
        print(" Tam giac cân")
    elif (a*a + b*b == c*c) or (a*a + c*c == b*b) or (b*b + c*c == a*a):
        print(" Tam giac vuong")
    else:
        print(" Tam giac thuong")

    chu_vi = a + b + c
    p = chu_vi / 2
    dien_tich = (p * (p - a) * (p - b) * (p - c))

    print("Chu vi =", chu_vi)
    print("Dien tich =", dien_tich)
```

```python
#Bài 5
a, b, c = map(float, input("Nhap vao 3 so: ").split())
delta = b*b - 4*a*c
if delta < 0:
    print("Phuong trinh vo nghiem.")
elif delta == 0:
    x = -b / (2*a)
    print("Phuong trinh co nghiem kep x =", x)
else:
    x1 = (-b + delta**0.5) / (2*a)
    x2 = (-b - delta**0.5) / (2*a)
    print("Phuong trinh co 2 nghiem phan biet:")
    print("x1 =", x1)
    print("x2 =", x2)
    
```
