```python
#bài 8
a, b = map(int, input("Nhap hai so nguyen duong a va b: ").split())
print("CCac uoc duong chung cua", a, "va", b, "la:")
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
```

```python
#bài 9
import math
a = int(input("Nhap so nguyen duong a: "))
k = int(math.sqrt(a))
if k * k == a:
    print(a, "la so chinh phuong")
else:
    print(a, "khong phai la so chinh phuong")
```

```python
#bài 10
a = int(input("Nhap so nguyen duong a: "))
tong = 0
for i in range(1, a):
    if a % i == 0:
        tong += i
if tong == a:
    print(a, "la so hoan hao")
else:
    print(a, "khong phai la so hoan hao")
```

```python
#bài 15
n = int(input("Nhap n: "))
count = 0

while n > 0:
    count += 1
    n //= 10

print(count)
```

```python
#bài 16
n = int(input("Nhap n: "))

even = 0
odd = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        even += 1
    else:
        odd += 1
    n //= 10

print("So chu so chan:", even)
print("So chu so le:", odd)
```

```python
#bài 17
n = int(input("Nhap n: "))

total = 0

while n > 0:
    digit = n % 10
    total += digit
    n //= 10

print("Tong cac chu so:", total)
```

```python
#bài 18
n = int(input("Nhap n: "))

if n < 1:
    print("Khong phai dang 3^k")
else:
    while n % 3 == 0:
        n //= 3

    if n == 1:
        print("La so dang 3^k")
    else:
        print("Khong phai dang 3^k")
```

```python
#bài 19
A = int(input("Nhap A: "))

f1 = 1
f2 = 1

if A < 1:
    print("Khong co so Fibonacci phu hop")
else:
    while f2 <= A:
        f1, f2 = f2, f1 + f2

    print("So Fibonacci lon nhat khong vuot qua A:", f1)
```
