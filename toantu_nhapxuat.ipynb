```python
#W2a3
a = int(input("Nhap so nguyen a: "))
b = int(input("Nhap so nguyen b: "))
tong = a + b
hieu = a - b
tich = a * b
phan_nguyen = a // b
phan_du = a % b
chia_thuc = a / b
print("Tong:", tong)
print("Hieu:", hieu)
print("Tich:", tich)
print("Phan nguyen:", phan_nguyen)
print("Phan du:", phan_du)
print("Chia thuc: {:.2f}".format(chia_thuc))
```
